```python
from nba_api.stats.endpoints import LeagueGameLog
import pandas as pd

# Get all games from the 2023-24 season
game_log = LeagueGameLog(season="2024-25", season_type_all_star="Regular Season")
data = game_log.get_data_frames()[0]  # Convert to DataFrame

# Save to CSV
data.to_csv("nba_game_logs_2023_24.csv", index=False)

# Display the first few rows
data.head()
```
