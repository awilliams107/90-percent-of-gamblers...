```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

```

```python
df = pd.read_csv('./data/lebron_game_data.csv')
df2 = pd.read_csv('./data/TEAM MISC STATS.csv')
df['Home'] = df['Unnamed: 5'].apply(lambda x: 1 if pd.isna(x) or x == '' else 0)

df.drop(columns="Unnamed: 5")
```

```python
df = df[df['MP'].notna() & (df['MP'] != '') & (df["MP"] != 'Inactive')]
print(df.columns)
```

```python
merged_df = pd.merge(df, df2, left_on='Opp', right_on='Team', how='left')
```

```python
merged_df = merged_df.drop(columns=['Rk', 'Unnamed: 7', 'GS', 'Unnamed: 5', 'Tm', 'Age', 'Date', 'G', '+/-','Team', 'Opp', 'W', 'L', 'PW','PL','MOV','SOS','SRS','ORtg','FTr', '3PAr', 'eFG%', 'TOV%', 'ORB%', 'FT/FGA'])
print(merged_df.columns)
```
